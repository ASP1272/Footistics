```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Load your DataFrame with player stats
player_df = pd.read_csv(r"E:\2nd year\3rd sem\Data science\Project\FInal_DF\am.csv")

# Remove the "Position" column from the DataFrame
player_df = player_df.drop("Position", axis=1)

# Choose a player for whom you want to find similar players
selected_player = "Bruno Fernandes"  # Replace with the name of the player you're interested in

# Extract the stats of the selected player from the DataFrame
selected_player_stats = player_df[player_df['Player'] == selected_player].drop('Player', axis=1)

# Calculate the similarity with other players using cosine similarity
similarity_scores = cosine_similarity(selected_player_stats, player_df.drop('Player', axis=1))
player_df['Cosine_Similarity'] = similarity_scores[0]  # Assign the similarity scores as a single row

# Sort the DataFrame by similarity in descending order to get the most similar players
similar_players_cs = player_df.sort_values(by='Cosine_Similarity', ascending=False)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors

# Load your DataFrame with player stats
player_df = pd.read_csv(r"E:\2nd year\3rd sem\Data science\Project\FInal_DF\am.csv")

# Remove the "Position" column from the DataFrame
player_df = player_df.drop("Position", axis=1)
player_df = player_df.drop("Unnamed: 0", axis=1)

# Load your DataFrame 'am_df' with player parameters

# Exclude non-numeric columns like player names
numeric_columns = player_df.select_dtypes(include=['number'])

# Check if you have any missing values and handle them if needed
numeric_columns.fillna(0, inplace=True)  # You can replace NaN values with zeros or handle them differently

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(numeric_columns)

# Choose the number of neighbors (k) - adjust this based on your preference
k = 5
knn_model = NearestNeighbors(n_neighbors=k, algorithm='auto', metric='euclidean')
knn_model.fit(X_scaled)

# Ask the user for the name of the player
player_name = input("Enter the name of the player you want to find similar players for: ")

# Find the index of the player with the given name
player_index = player_df[player_df['Player'] == player_name].index[0]

# Replace 'query_player_index' with the index of the player you want to find similar players for
query_player = [X_scaled[player_index]]
distances, indices = knn_model.kneighbors(query_player)

# 'indices' contains the indices of similar players, and 'distances' contains their distances
similar_player_indices = indices[0]
similar_player_distances = distances[0]

# You can then use these indices to retrieve the similar players from your original DataFrame
similar_player_knn = player_df.iloc[similar_player_indices]
```
